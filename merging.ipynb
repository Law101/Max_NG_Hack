```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('Train.csv')
```

```python
train.head()
```

```python
train['service_type'] = train['Personal or Business'].map({'Business':'Delivery','Personal':'Hailing'})
```

```python
train.head()
```

```python
train = train.drop(columns=['Vehicle Type','Platform Type','Temperature','Precipitation in millimeters','Personal or Business'],inplace=False)
```

```python
train.to_csv('train_bi.csv',index=False)
```

```python
train['Placement - Time']
train['Placement_time'] = pd.to_datetime(train['Placement - Time'], format='%I:%M:%S %p')
train['Placement_hour'] = train.Placement_time.apply(lambda x: x.hour)
```

```python
train.info()
```

```python
train.head()
```

```python
import folium
```

```python
def generateBaseMap(default_location=[-1.317755, 36.830370], default_zoom_start=11):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
from folium.plugins import HeatMap
train_c = train.copy()
train_c['count'] = 1
base_map = generateBaseMap()
HeatMap(data=train_c[['Pickup Lat', 'Pickup Long', 'count']].groupby(['Pickup Lat', 'Pickup Long']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(base_map)
base_map
```

```python
base_map.add_child(folium.ClickForMarker(popup='Potential Location'))
```

```python
train.head()
```

```python
place_hour_list = []
for hour in train_c.Placement_hour.sort_values().unique():
    place_hour_list.append(train_c.loc[train_c.Placement_hour == hour, ['Pickup Lat', 'Pickup Long', 'count']].groupby(['Pickup Lat', 'Pickup Long']).sum().reset_index().values.tolist())
```

```python
from folium.plugins import HeatMapWithTime
base_map = generateBaseMap(default_zoom_start=11)
HeatMapWithTime(place_hour_list, radius=5, gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
base_map
```
